- Due to the limited availability of labelled datasets in the field, I have had to use a mixture of datasets to achieve the intended result.

## Dataset 1 (The Singapore Jobs Dataset)

- This dataset is a collection of jobs extratcted from the official Singapore Government job board. 
- It contains all nature of jobs ranging from IT to Construction. 
- It also has skills required with every listing.  o

```python
import pandas as pd
```

```python
# load dataset
singapore_data = pd.read_csv('../data/singaporedataset.csv')
singapore_data.sample(10)
```

```python
singapore_data.shape
```

```python
# dropping unneeded cols



singapore_df = singapore_data.copy()

singapore_df = singapore_data.drop(columns=['jobs/seniority','jobs/company_info','jobs/company_name','jobs/job_category/1','jobs/location','jobs/salary','jobs/posting_date','jobs/expiry_date','jobs/no_of_applications','jobs/job_id', 'jobs/job_category/2','jobs/job_category/3', 'jobs/job_category/4', 'jobs/job_category/5', 'jobs/job_category/6',])

singapore_df.sample(20)
```

```python
#  Getting all unique values from job category to aid in selecting IT related roles.
unique = singapore_df['jobs/job_category/0'].unique()
unique
```

```python
# Filtering by IT jobs alone

singapore_it_df = singapore_df[singapore_df['jobs/job_category/0'] == 'Information Technology']
singapore_it_df
```

```python
print(f"The filtered dataset contains {singapore_it_df.shape[0]} rows and {singapore_it_df.shape[1]} columns")
```

#### Combining skills into one target column of strings

```python
singapore_it_df['skills'] = singapore_it_df[singapore_it_df.columns[4:17]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1
)
singapore_it_df['job_description'] = singapore_it_df["jobs/requirements_and_role"]

singapore_it_df.sample(5)
```

```python
finalDF = singapore_it_df.drop(singapore_it_df.columns[1:-2],axis=1)
finalDF.sample(1)
```

```python
# Export final Dataset

finalDF.to_csv("dataset1.csv")
```

# Dataset 2

- The Singapore Dataset was not enough to use as after filtering unwanted roles, we ended up with 3800 rows, which may be enough to train a BERT model, but we can get more. 
- This dataset was used by Aishwarya Pimplikar, Rajashree Pati, Amanpreet Singh, Arya Shah in their paper [CareerMap- Mapping Tech Roles With Personality & Skills](https://data.mendeley.com/datasets/5z68cvxssn/1)
- It is a dataset of skills and their relevance to labelled roles. This data can be augmented to be skills required for roles. 
- They used a labelling system where they ranked relevancy of skills to a role on a scale of 1 - 7, where 1 represented 'Not Interested'(not needed) and 7 represented a skill that would greatly improve your suitability to the role.
- The dataset also has soft skills that I will not be interested in. 
- I took anything above level 5 to be a required skill.  

```python
# load data

careermap_data = pd.read_csv('../data/dataset2.csv')
careermap_data.sample(10)
```

```python
# Remove unwanted rows

careermap_df = careermap_data.copy()

careermap_df = careermap_data.drop(columns=['Openness', 'Extraversion', 'Agreeableness', 'Emotional_Range', 'Conversation', 'Openness to Change', 'Hedonism', 'Self-enhancement', 'Self-transcendence', 'Conscientousness'])

careermap_df.sample(20)
```
